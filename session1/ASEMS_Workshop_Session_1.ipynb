## OpenCV

OpenCV is another useful package which is employed in image analysis. There are built-in functionalities for many different image processing and computer vision tasks in OpenCV. In this section we will discuss some of these interesting operations. 

To begin with, let's see how we read an image and visualize it:

```python
import cv2
from google.colab.patches import cv2_imshow

# download image from the web (picsum for random image)
! wget -O "img.png" "https://picsum.photos/400/200"

# read image from the file system
image = cv2.imread("img.png")

# print the image size
print(image.shape)

# visualizing the image
# google colab prevents us from using cv2.imshow()
cv2_imshow(image)
```

Some notes:
* Images are stored as 2 dimensional arrays in computer. 
* 1 spatial dimension and 1 color/spectral. 
* size of channel dimension is often 1 (grayscale) or 3 (RGB/colored)

Let's separate these channels and visualize each using the things we have learned so far:

image.png

```python
# slicing channels (notice anything weird?!)
red_channel = image[:,:,2]
green_channel = image[:,:,1]
blue_channel = image[:,:,0]

print("Red channel (as grayscale)")
cv2_imshow(red_channel)

print("Green channel (as grayscale)")
cv2_imshow(green_channel)

print("Blue channel (as grayscale)")
cv2_imshow(blue_channel)
```

How to see each channel in its related color? 

* use "merge" function from opencv

```python
# create an empty array to represent the empty channels
# note the type
zero_2d = np.zeros(red_channel.shape).astype('uint8')

# merging the single channels with the empty channel to produce the color effect
red_channel = cv2.merge([zero_2d,zero_2d,red_channel])
green_channel = cv2.merge([zero_2d,green_channel,zero_2d])
blue_channel = cv2.merge([blue_channel,zero_2d,zero_2d])


print("Red channel (as grayscale)")
cv2_imshow(red_channel)

print("Green channel (as grayscale)")
cv2_imshow(green_channel)

print("Blue channel (as grayscale)")
cv2_imshow(blue_channel)
```

Now let's calculated an index called Triangular Vegetation Index (TGI). This index which is in form of a grayscale image shows what parts of an image has more vegatation. The formula is:

TGI = -0.39*R-0.61*B+G

https://giscrack.com/list-of-spectral-indices-for-sentinel-and-landsat/

```python
# download image from the web (picsum for random image)
! wget -O "veg_img.png" "https://picsum.photos/id/1025/400/400"

# read image from the file system
veg_image = cv2.imread("veg_img.png")

# slicing channels
red_channel = veg_image[:,:,2]
green_channel = veg_image[:,:,1]
blue_channel = veg_image[:,:,0]

TGI = (-0.39*red_channel-0.61*blue_channel+green_channel)

cv2_imshow(veg_image)
cv2_imshow(TGI)
```

## Json

Json is another convenient format to store data. It has been widely used in many areas from Mobile App Developmenet, Web Development to scientific computing and machine learning. It is based on the concept of key and value lists. To understand this, let's begin with defining a dictionary (pairs of key and value) in python:

```python
name_dictionary = {'person1':5209999999,'person2':9280003333}
print(name_dictionary)
```

```python
keys = name_dictionary.keys()
print(keys)
```

```python
name_dictionary['person2']
```

```python
name_dictionary['person3'] = 3048882233
print(name_dictionary)
```

```python
# Caller ID
id_lookup_dictionary = {}

# adding multiple using for loop
for key in keys:
  phone = name_dictionary[key]
  id_lookup_dictionary[phone] = key

print(id_lookup_dictionary)
```

```python
# check who is calling
if 5209999999 in id_lookup_dictionary:
  print(id_lookup_dictionary[5209999999])
else:
  print("New phone number")
```

Json is simply a dictionary or list of dictionaries or any nested combinatios of lists and dictionaries. We can save python dictionaries/lists of dictionaries into a Json file and read it back using the following set of commands:

```python
import json

# creating a json file
with open("id_lookup_file.json","w") as f:
  json.dump(id_lookup_dictionary,f)

! ls

# loading data from json file
with open("id_lookup_file.json","r") as f:
  new_dict = json.load(f)

# printing the loaded dictionary from json
print(new_dict)
```

# Loading and Visualizing Data

Data exploration is an essential skill thoughout STEM. Lets learn to open and visualize data from a comma separated value file (csv) using three Python packages: Pandas, Seaborn, and Plotly Express.

Below a link to a csv file we have saved for you, but this could also be the path to a file on your computer!


```python
# Saving the filepath to our data as the variable data_filepath
data_filepath = 'https://data.cyverse.org/dav-anon/iplant/projects/phytooracle/season_10_lettuce_yr_2020/level_2/stereoTop/season10_plant_detection/2020-03-03_detection.csv'
```

By reading the dot notation below we can see that we are calling a function from the pandas library (pd) called 'read_csv'. We can also see that we are using data_filepath (defined above) as the only argument.

Finally, we are storing the results in the variable 'df'.

This function will take in a path to a csv file and read it into a table called a dataframe (hence the variable being named df). A DataFrame is a special table that can be created and manipulated by pandas!
